## 2. 기본적인 함수 이용

```python
from konlpy.tag import Hannanum
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Komoran
from eunjeon import Mecab 
```

- Hannannum(한나눔) KAIST 말뭉치, KAIST Semantic Web Research Center 개발
- Okt(Open Korean Text) - 오픈소스 한국어 분석기, (Twitter)
- Kkma(꼬꼬마): 세종 말뭉치를 이용해 생성된 사전, 서울대 연구실에서 개발.
- Komoran : Java로 쓰여진 오픈 소스 한글 형태소 분석기
- Mecab : 일본어 형태소 분석기를 한국어에 맞도록 수정함.

```python
# 함수를 변수에 저장
han = Hannanum()
okt =  Okt()
kkma = Kkma()
kom = Komoran()
mecab = Mecab()
```

```python
# 예시 문장
sentence = "구글 뉴스는 자동 집계 알고리즘에 의해 수천 곳의 발행사로부터 최신의 정보를 선별한다."
```

### 형태소 추출

- 텍스트에서 형태소를 추출해 반환한다.

- .morphs

```python
#Hannanum 사용.
han.morphs(sentence)
```

```python
#Komoran사용.
kom.morphs(sentence)
```

### 명사 추출

- 텍스트에서 명사를 추출해 반환한다.

- .nouns()

```python
han.nouns(sentence)
```

```python
okt.nouns(sentence)
```

```python
kkma.nouns(sentence)
```

### 품사 부착

- 텍스트에서 품사를 부착해 반환한다.

- .pos()

```python
kom.pos(sentence)
```

```python
mecab.pos(sentence)
```

## 3. 연습

### 뉴스 크롤링

```python
from bs4 import BeautifulSoup
from urllib import request, parse
```

```python
url = "https://news.v.daum.net/v/20200926215958649"
res = request.urlopen(url)
soup = BeautifulSoup(res, "html.parser")
a_list = soup.select("#harmonyContainer > section > p")
```

```python
# list 형태로 기사 변환해 저장.
x = []
for a in a_list:
    name = a.string
    x.append(name)
```

```python
#list 형태 하나로 합치기.
news = ""
for i in range(1,21): news += x[i]
```

```python
news
```

### konlpy로 분석

```python
han.morphs(news)
```

```python
mecab.nouns(news)
```

```python
mecab.pos(news)
```
